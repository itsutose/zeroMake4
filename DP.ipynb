```python
# 反復方策評価 100回行うアルゴリズム

V = {'L1' : 0.0, 'L2' : 0.0}
new_V = V.copy()

for _ in range(100):
    new_V['L1'] = 0.5 * (-1 + 0.9 * V['L1']) + 0.5 * (1 + 0.9 * V['L2'])
    new_V['L2'] = 0.5 * (0 + 0.9 * V['L1']) + 0.5 * (-1 + 0.9 * V['L2'])
    V = new_V.copy()
    print(_, V)
```

```python
V = {'L1' : 0.0, 'L2' : 0.0}
new_V = V.copy()
cnt = 0


while True:
    new_V['L1'] = 0.5 * (-1 + 0.9 * V['L1']) + 0.5 * (1 + 0.9 * V['L2'])
    new_V['L2'] = 0.5 * (0 + 0.9 * V['L1']) + 0.5 * (-1 + 0.9 * V['L2'])
    
    delta = abs(new_V['L1'] - V['L1'])
    delta = max(delta, abs(new_V['L2'] - V['L2']))

    V = new_V.copy()
    # print(_, V)

    cnt += 1
    if delta < 0.0001:
        print(V)
        print(cnt)
        break
```
