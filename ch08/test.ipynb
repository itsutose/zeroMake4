```python
import numpy as np
import gym
import time

env = gym.make('CartPole-v0',render_mode='human')
state = env.reset()
done = False


time.sleep(0.5)
while not done:
    time.sleep(0.05)
    env.render()
    action = np.random.choice([0, 1])
    next_state, reward, done, info, info1 = env.step(action)
env.close()
```

```python
import numpy as np
import gym

env = gym.make('CartPole-v0', render_mode = 'human')    # Cartpole定義
env.reset()    # Cartpoleの状態初期化

for i in range(1000):
    env.render()    # Cartpoleのアニメーション
    observation, reward, done, info, info1 = env.step(env.action_space.sample())  # カートを動かし、結果を返す
    print("Step:",i,done,"Reward:",reward,"Obs:",observation)
```
